```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_excel("moni6.xlsx")
df=pd.DataFrame(data)
print(df)
grouped_by=df.groupby("date")["co"].sum()
print(grouped_by)
plt.bar(grouped_by.index,grouped_by.values)
plt.xlabel("date",fontsize=20)
plt.ylabel("co",fontsize=20)
plt.title("Air Quality",fontsize=20)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_excel("moni6.xlsx")
df=pd.DataFrame(data)
print(df)
plt.pie(df["pm2_5"],labels=df["date"])
plt.xlabel("pm2_5",fontsize=20)
plt.ylabel("date",fontsize=20)
plt.title("Air Quality",fontsize=20)
plt.show()
```
